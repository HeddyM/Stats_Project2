# Yelp

Send a request to Yelp with a small radius (1000m) for all the bike stations in your city of choice. 

```python
YLPAPI_KEY = os.getenv('YELP_API_KEY')
```

```python
bikes_df = pd.read_csv('bikes_data.csv')
bikes_df.head()
```

```python
api_key = YLPAPI_KEY
```

```python
def yelp_bars_fs(latitude, longitude, api_key):
    url = f"https://api.yelp.com/v3/businesses/search?latitude={latitude}&longitude={longitude}&term=%22bar%22&radius=1000&limit=50"
    headers = {
    "accept": "application/json",
    "Authorization": f"Bearer {api_key}"
    }
    response = requests.get(url, headers=headers)
    return response.json() 
```

```python
# bikes_df = bikes_df.head()
# bikes_df
```

```python
df_concatYLP = pd.DataFrame()

for bk_station in range(len(bikes_df)):
    latitude = bikes_df.at[bk_station, 'latitude']
    longitude = bikes_df.at[bk_station, 'longitude']
    
    df_resultsYLP = yelp_bars_fs(latitude, longitude, api_key)
    df_resultsYLP = pd.DataFrame(df_resultsYLP['businesses'])
    df_resultsYLP['latitude'] = latitude
    df_resultsYLP['longitude'] = longitude
    
    df_concatYLP = pd.concat([df_concatYLP, df_resultsYLP], axis=0)

df_concatYLP.reset_index(drop=True, inplace=True)
```

```python
# full data from Yelp in a dataframe
df_concatYLP
```

```python
# creating a smaller dataframe with the columns that I need
col_filter = ['name', 'review_count', 'rating', 'distance', 'price', 'latitude', 'longitude']
df_finalYLP = df_concatYLP[col_filter]
df_finalYLP
```

```python
df_finalYLP.to_csv('Yelp_result_details.csv', index=False)
```
