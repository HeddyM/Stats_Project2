Join the data from Part 1 with the data from Part 2 to create a new dataframe.

```python
import pandas as pd
```

```python
# putting bikes data saved from Part 1 into a dataframe
bikes_df = pd.read_csv('bikes_data.csv')
bikes_df.head()
```

```python
# confirming data types of the bikes_df columns
bikes_df.dtypes
```

```python
# putting selected foursquare data saved from Part 2 into a dataframe
# also changing columns names so there is no duplicate columns names between foursquare and yelp files when merged
new_column_names = ['ll', 'edu_facility', 'edu_distance']

foursquare_df = pd.read_csv('FSq_result_details.csv', header=None, names=new_column_names, skiprows=1)
foursquare_df.head()
```

```python
# confirming data types of the foursquare_df columns
foursquare_df.dtypes
```

```python
# putting selected yelp data saved from Part 2 into a dataframe
# also changing columns names for clarity in final merged dateframe
new_column_names = ['venue_name', 'venue_review_count', 'venue_rating', 'venue_distance', 'venue_price', 'latitude', 'longitude']

yelp_df = pd.read_csv('yelp_result_details.csv', header=None, names=new_column_names, skiprows=1)
yelp_df.head()
```

```python
# confirming data types of the yelp_df columns
yelp_df.dtypes
```

```python
# merge bike data from Part 1 and foursquare data from Part 2 into a new dataframe on 'll'

first_merge_df = pd.merge(bikes_df, foursquare_df, on='ll')

first_merge_df.head()
```

```python
# merge new dataframe and yelp data from part 2 into final dataframe on 'latitude' and longitude'

stats_project_df = pd.merge(first_merge_df, yelp_df, on=['latitude', 'longitude'])

stats_project_df.head()
```

```python
# saving the data in a csv file just to be safe
stats_project_df.to_csv('Project_data_all.csv', index=False)
```

```python
# confirming the data types of the stats_project.df
stats_project_df.dtypes
```

Provide a visualization that you used as part of your EDA process. Explain the initial pattern or relationship you discoved through this visualization. 

```python

import matplotlib.pyplot as plt
```

```python
# creating a histrogram showing the size of each bike station 
# first need to drop duplicates as each bike station appears multiple times in the dataframe

each_station = stats_project_df.drop_duplicates(subset='uid')['slots']

plt.hist(each_station)
plt.title('Bike Station Size in Toronto')
plt.xlabel('Number of Slots at a Station')
plt.ylabel('Number of Stations of that Size')

plt.show()
```

### Visualization 1 - Size of Bike Stations in Toronto
#### In this visualization I could see that the bike stations in Toronto are not of uniform size (as determined by number of bike slots at a station). Most of the stations have 15-20 bike slots but there are some with only 10 and some with as many as 50.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Plot a scatter plot between number of bike slots at each station and number bikes available. 

# first I need to ensure that I filter by bike station to remove duplicates

station_use = stats_project_df.drop_duplicates(subset='uid')
```

```python
sns.relplot(x = "slots", y = "free_bikes", data = station_use)
plt.title('Bike Station Size by Number of Available Bikes on one day in Toronto')
plt.xlabel('Station size (Number of bike slots)')
plt.ylabel('Number of bikes available')
```

### Visualization 2 - Number of available Bikes on he day when the data was pulled from city bikes
#### In this visualization I was interested in understanding if there was any pattern in the number of free bikes at a station based on its size. In other words are some stations under utilized and have many free bikes available and others that have mosty emply slots. You can see that there are some stations with over 30 bikes slots and no free bikes available while other stations with over 40+ bike slots and 40+ bikes available. This may be a function of the date, time of day and location but analyzing this type of data over a longer time period would be useful to determine if the bike stations are sized appropriately for the demand of the location or if changes to the mix need to be made to better serve Torontonians.

```python
# now I am interested in the number of bike stations near each major Toronto College and University

# determined which colleges and universities to include in the analysis
array = ['University of Toronto', 'Ryerson University', 'George Brown College', 'York University', 'Humber College', 'Centennial College']

# returned the dataframe with only the columns of interest
col_filter = ['uid', 'station', 'slots', 'edu_facility']
df_edu_stations = stats_project_df[col_filter]

# filtered that data frame based on the array created above and removed duplicates
smaller_set = df_edu_stations[df_edu_stations['edu_facility'].isin(array)].drop_duplicates()

plt.figure(figsize=(12, 6))
sns.countplot(data=smaller_set, x='edu_facility')
plt.xlabel('Post_Secondary Institution')
plt.ylabel('Number of Bike Stations within 1000 metres')
plt.title('Number of Bikes Stations in Toronto Near Major Post_Secondary Institutions')
```

```python
# I could also graph number of total number of bike slots (available bikes) near these institutions

# code to get sum of bike slots and sort so graph would present in order of bar height
bike_counts = smaller_set.groupby('edu_facility')['slots'].sum().reset_index()
bike_counts_sorted = bike_counts.sort_values(by='slots', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(data=bike_counts_sorted, x='edu_facility', y='slots')
plt.xlabel('Post_Secondary Institution')
plt.ylabel('Number of Bike Slots within 1000 metres')
plt.title('Number of Bikes Slots (available bikes) in Toronto Near Major Post_Secondary Institutions')

plt.xticks(rotation=45)
plt.show
```

### Visualizations 3 and 4 - Number of Bike stations and available bikes near major post-secondary institutions
#### In these two visualizations I was interested in understanding the availability of bike rentals near major post-secondary institutions in Toronto. The result is not surprising given the location of these institutions. Although access to rental bikes might be attractive to students, staff and faculty it is clear that the farther away the institution is from the downtown core the less availability of these rental bikes. York University has more students than Toronto Metropolitaton University (formerly Ryerson University) but has less than half of the available rental bike slots within 1000 metres of its campus.
